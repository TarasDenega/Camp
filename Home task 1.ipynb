```python
import numpy as np
import matplotlib.pyplot as plt

# Ваш випадковий масив
np.random.seed(2000)
a = np.random.randint(1, 100, 20)

# Знаходимо початок і кінець ями
start_of_descent = None
end_of_descent = None

# Знаходимо найглибшу точку і її індекс
deepest_point_index = np.argmin(a)
deepest_point = a[deepest_point_index]

# Шукаємо початок і кінець спадаючої послідовності в ямі навколо найглибшої точки
for i in range(deepest_point_index, 0, -1):
    if a[i] > a[i - 1]:
        start_of_descent = i
        break

for i in range(deepest_point_index, len(a) - 1):
    if a[i] > a[i + 1]:
        end_of_descent = i
        break

# Малюємо графік
x = np.arange(len(a))
plt.plot(x, a)
plt.title('Випадковий графік')
plt.xlabel('X-вісь')
plt.ylabel('Y-вісь')

# Визначаємо ширину виділеної зони для найглибшої точки
span_width = 1.0  # Початкова ширина виділеної зони

if deepest_point_index > 0 and deepest_point < a[deepest_point_index - 1]:
    span_width = 2.0  # Збільшуємо ширину, якщо точка глибша

if deepest_point_index < len(a) - 1 and deepest_point < a[deepest_point_index + 1]:
    span_width = 2.0  # Збільшуємо ширину, якщо точка глибша

# Виділяємо зону навколо найглибшої точки червоною зоною
plt.axvspan(deepest_point_index - span_width / 2, deepest_point_index + span_width / 2, facecolor='red', alpha=0.3)

# Виділяємо зону ями жовтою зоною
if start_of_descent is not None and end_of_descent is not None:
    plt.axvspan(start_of_descent, end_of_descent, facecolor='yellow', alpha=0.3)

plt.show()

print(f'Найглибша точка ({deepest_point}).')
```
